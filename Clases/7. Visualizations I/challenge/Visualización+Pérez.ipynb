```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

# 1. Cargar este archivo en Python. Realizar estadísticas descriptivas básicas

```python
df = pd.read_csv('HRDataset_v14.csv')
print("Descripción de los datos:\n", df.describe(), "\n")
print("Información de los datos: \n")
print(df.info())
```

# 2. Realizar un histograma con los salarios. Qué rango(s) de salarios son los más populares?

```python
plt.figure(figsize= (60,45)) #no me cambia el tamaño por más de que le modifique estos valores. Por que podría ser?
ax = sns.displot(data=df, x='Salary', kind='hist', fill=True) 
ax.set(xlabel='Salarios', ylabel='Cantidad', title='Salarios', xlim=[45000, 200000])
plt.show()
```

```python
print ("Los salarios más frecuentes se encuentran rondando los $65000 ")
```

# 3. Realizar un gráfico de violín con los salarios, discriminados por género. 

```python
fg = sns.violinplot(data=df, x='Sex', y='Salary')
plt.xlabel('Sexo')
plt.ylabel('Salario')
```
